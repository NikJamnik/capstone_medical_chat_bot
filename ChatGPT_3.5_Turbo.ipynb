**1.Load PDF File and make chunks**

```python
# Load the PDF
loader = PyPDFLoader("/content/drive/MyDrive/Capstone/Data/ClinicalDermatology.pdf")
pages = loader.load()

# Split into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = splitter.split_documents(pages)
```

```python
from langchain_pinecone import PineconeVectorStore
from google.colab import userdata
# Step 1: Connect to your existing index
pc = Pinecone(api_key=pinecone_key)
vectorstore = PineconeVectorStore(
    index_name="dermatologist-v2",
    embedding=embedding,
    namespace="clinical-derm"
)

# Step 2: Upload in batches (100–200 chunks per batch)
from tqdm import tqdm
batch_size = 100

for i in tqdm(range(0, len(chunks), batch_size)):
    chunk_batch = chunks[i:i+batch_size]
    vectorstore.add_documents(documents=chunk_batch)
```

```python

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0,
                 openai_api_key=openai_key)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    return_source_documents=True
)
```

```python
from ragas.metrics import (
    answer_relevancy,
    context_precision,
    context_recall,
    faithfulness,
)
from ragas import evaluate
from datasets import Dataset

ragas_data = []

# Step 1: Extract answers using RAG model
for question, ground_truth in manual_ground_truths.items():
    result = qa.invoke({"query": f"{question}"})
    answer = result["result"]

    contexts = [doc.page_content for doc in result.get("source_documents", [])]

    ragas_data.append({
        "question": question,
        "answer": answer,
        "contexts": contexts,
        "ground_truth": ground_truth
    })

# Step 2: Convert to HuggingFace Dataset
dataset = Dataset.from_list(ragas_data)

# Step 3: Evaluate with RAGAS
results = evaluate(
    dataset,
    metrics=[
        answer_relevancy,
        context_precision,
        context_recall,
        faithfulness,
    ]
)

# Step 4: Print results
print(results)
```

```python
def evaluate_with_tfidf_precision_faithfulness(ragas_data, top_k: int = None, threshold: float = 0.3):
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import cosine_similarity
    import numpy as np
    import pandas as pd

    precision_list = []
    faithfulness_list = []

    for sample in ragas_data:
        truth = sample['ground_truth']
        contexts = sample["contexts"]
        answer = sample.get("answer", "")  # optional answer for faithfulness

        if top_k:
            contexts = contexts[:top_k]

        # === CONTEXT PRECISION ===
        documents = [truth] + contexts
        tfidf = TfidfVectorizer()
        tfidf_matrix = tfidf.fit_transform(documents)
        truth_vec = tfidf_matrix[0]
        context_vecs = tfidf_matrix[1:]
        similarities = cosine_similarity(truth_vec, context_vecs).flatten()

        # Count contexts with similarity above threshold
        relevant_count = (similarities >= threshold).sum()
        precision = relevant_count / len(contexts) if contexts else 0
        precision_list.append(precision)

        # === FAITHFULNESS (Answer vs Contexts) ===
        if answer.strip():
            all_texts = contexts + [answer]
            tfidf_ans = TfidfVectorizer()
            tfidf_matrix_ans = tfidf_ans.fit_transform(all_texts)
            ans_vec = tfidf_matrix_ans[-1]
            context_vecs_ans = tfidf_matrix_ans[:-1]
            faithfulness_score = cosine_similarity(ans_vec, context_vecs_ans).max()
        else:
            faithfulness_score = np.nan  # if no answer available
        faithfulness_list.append(faithfulness_score)

    return {
        "avg_context_precision": np.mean(precision_list),
        "avg_faithfulness": np.nanmean(faithfulness_list),  # ignore missing answers
        "precision_scores": precision_list,
        "faithfulness_scores": faithfulness_list
    }
```

```python
scores = evaluate_with_tfidf_precision_faithfulness(ragas_data)
print("Average Context Precision:", scores["avg_context_precision"])
print("Average Faithfulness:", scores["avg_faithfulness"])
```

```python
# Extract contexts and queries from ragas_data
contexts = [doc for sample in ragas_data for doc in sample["contexts"]]
queries = [sample["question"] for sample in ragas_data]

# Optional: deduplicate contexts and queries (usually queries are unique)
unique_contexts = list(set(contexts))
unique_queries = list(set(queries))

# Create labels to keep track of origin
labels_contexts = ["context"] * len(unique_contexts)
labels_queries = ["query"] * len(unique_queries)

# Combine data for embedding
combined_text = unique_contexts + unique_queries
combined_labels = labels_contexts + labels_queries

# Embed using SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(combined_text, show_progress_bar=True)

# Reduce to 2D with UMAP
reducer = umap.UMAP(n_neighbors=10, min_dist=0.1, metric="cosine")
embedding_2d = reducer.fit_transform(embeddings)

# Create a DataFrame for plotting
df_umap = pd.DataFrame(embedding_2d, columns=["x", "y"])
df_umap["text"] = combined_text
df_umap["type"] = combined_labels

# Plot with different colors for queries and contexts
plt.figure(figsize=(12, 10))
for label, color in zip(["context", "query"], ["blue", "red"]):
    subset = df_umap[df_umap["type"] == label]
    plt.scatter(subset["x"], subset["y"], s=30, alpha=0.7, label=label, c=color)

# Annotate a few points (optional: just queries or both)
for i, row in df_umap[df_umap["type"] == "query"].head(15).iterrows():
    plt.annotate(row["text"][:40] + "...", (row["x"], row["y"]), fontsize=9, color='darkred')

plt.title("UMAP Projection of Queries and Retrieved Contexts")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create or open the CSV file
csv_file = "retrieval_latency_log.csv"

# Create header if file does not exist
try:
    with open(csv_file, "x", newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["timestamp", "query", "retrieval_latency"])
except FileExistsError:
    pass  # File already exists, skip header

# List of test queries
queries = [
    "What are the main causes of psoriasis?",
    "What are the main causes of acne and how to cure?",
    "What is the treatment for eczema?",
    "Explain the difference between a nodule and a tumor.",
    "What causes hair to turn grey early?",
    "How is a potential fungal infection tested?",
    "Is dapsone an effective treatment for leprosy?",
     "Which drugs can cause toxic reactive erythema?",
     "How many layers does skin have?"

]

# Run queries and log latency
for query in queries:
    start_time = time.time()
    result = qa.invoke({"query": query})
    latency = time.time() - start_time

    # Log to CSV
    with open(csv_file, "a", newline='') as f:
        writer = csv.writer(f)
        writer.writerow([datetime.now(), query, round(latency, 3)])

    print(f"Query: {query}")
    print("Answer:", result['result'])
    print(f"Retrieval Latency: {latency:.3f} seconds\n")
```
